```python
import pandas as pd
import torch
```

```python
path = r"D:\project\STUDY\NAM_2\TERM_2\DSA++\BaiTap\Computational Graph\TRAIN2(Sheet1).csv"

df = pd.read_csv(path, sep=";", decimal=",")

df = df.dropna(axis=1, how="all")

df.columns = df.columns.str.strip()

X = torch.tensor(df["midterm"].values, dtype=torch.float32)
X = X.reshape(-1, 1)

Y = torch.tensor(df["final"].values, dtype=torch.float32)
Y = Y.reshape(-1, 1)

W = torch.tensor([[0.0]], dtype=torch.float32, requires_grad=True)
b = torch.tensor([[0.0]], dtype=torch.float32, requires_grad=True)
```

```python


def train_linear_regression(X, Y, learning_rate=0.01, epochs=5000):
    W = torch.tensor([[0.0]], dtype=torch.float32, requires_grad=True)
    b = torch.tensor([[0.0]], dtype=torch.float32, requires_grad=True)

    for epoch in range(epochs):
        Y_pred = torch.matmul(X, W) + b

        loss = ((Y - Y_pred) ** 2).mean()

        loss.backward()

        with torch.no_grad():
            W -= learning_rate * W.grad
            b -= learning_rate * b.grad

        W.grad.zero_()
        b.grad.zero_()

    return W, b

def predict(midterm_score, W, b):
    return W.item() * midterm_score + b.item()
```

```python

W, b = train_linear_regression(X, Y, learning_rate=0.01, epochs=5000)

print("w =", W.item())
print("b =", b.item())
```
